Bootstrap이란 ? 

    css 프론트엔드 프레임워크(Toolkit) --> 미리 만들어진 다양한 디자인 요소들을 제공하여 웹 사이트를 빠르고 쉽게 개발할 수 있도록 한다.

### bootstrap 기본 사용법

- Bootstrap의 모든 것들은 클래스로 만들어져 있다.

1. m{}-* : margin top - 값(pixel이 아닌 부트스트랩이 정해놓은 값)

    {m : roperty}{t : sides} - {numb : size}

-- Bootstrap에서 클래스 이름으로 spacing을 표현하는 방법



! Bootstrap에는 특정한 규칙이 있는 클래스 이름으로 이미 스타일 레이아웃이 작성되어 있다.

### Typography : 제목, 본문, 텍스트, 목록 등

1. Display Headings : 기존 heading(h1 ~ 6)보다 더 눈에 띄는 제목이 필요할 경우

    - 기본적으로 h1 태그 위에 쓴다
    
    - "display-1~6"

2. Inline text element

3. Lists

### Colors : Text, Border, Background 및 다양한 요소에 사용되는 Bootstrap의 색상 키워드

- 제공되는 색상과 bootstrap의 색상은 살짝 다르다.

1. .text-* : 글씨의 색상을 바꾸는 코드. *에 색상이 들어간다.

2. .bg-* : 글자의 배경색상을 바꾸는 코드.

3. border-* : box 경계

### Component : Bootstrap에서 제공하는 UI 관련 요소

: 일관된 디자인을 제공하여 웹사이트의 구성 요소를 구축하는 데 유용하게 사용

- Alerts

    - alert alert-*

- Badges

    - badge bg-*

- Buttons

    - `<button type="button" class="btn">Base class</button>`

    - 다른 태그에도 사용이 가능하다.

- Cards : 자주사용된다. 공식문서 참고

- Navbar

    - navbar navbar-*

- Carousel

    - carousel slide

    - carousel 사용 시 carousel id와 button의 target을 잘 확인해야한다(독립적이여야 함)

- Modal

    - carousel과 마찬가지로 modal id와 button의 target의 잘 확인하자

```
- carousel과 modal을 코드 내부에 작성할 필요가 없다.

- 하단에 구조를 따로 빼주어도 무방함

- button만 내부에 있으면 된다.
```

### OOCSS

: 객체 지향적 접근법을 적용하여 css를 구성하는 방법론

- CSS를 효율적이고 유지 보수가 용이하게 작성하기 위한 일련의 가이드

1. 구조와 스킨을 분리

    - 재사용 가능성 향상

2. 컨테이너와 콘텐츠를 분리

    - 모든 버튼의 공통 구조를 정의 + 각각의 스킨을 정의

    - 객체에 직접 적용하는 대신 객체를 둘러싸는 컨테이너에 스타일을 적용

    - 스타일을 정의할 때 위치에 의존적인 스타일을 사용하지 않도록 함

    - 콘텐츠를 다른 컨테이너로 이동시키거나 재배치할 때 스타일이 깨지는 것을 방지
