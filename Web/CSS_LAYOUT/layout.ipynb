- Box 구성요소

    - Margin : 이 박스와 *다른 요소 사이*의 공백. 가장 바깥쪽 영역

    - Border : 콘텐츠와 패딩을 감싸는 테두리 영역

    - Padding : *콘텐츠 주위*에 위치하는 공백 영역

    - Content : 콘텐츠가 표시되는 영역



! 페이지에서 먼저 테스트해본 뒤 적용하는 것도 방법이다.

<br>

- 너비와 높이는 height & width로 나타냄

    - 지정되는 요소의 너비와 높이는 콘텐츠 영역을 대상으로 한다.

    - ==> 레이아웃 전체의 width는 더 커질 수 있다.

### 박스 타입

Block & Inline

Normal flow : css를 적용하지 않았을 경우 웹페이지 요소가 기본적으로 배치되는 방향 : 왼쪽 위

<br>

##### Inline Content : 자기 자신이 필요한 영역만 차지한다. (좌-우 방향)

    - a, span, img

    - 새로운 행으로 나뉘지 않음

    - !!width, height 속성을 사용할 수 없다!!(Content 크기에 따라 결정되기 때문)

    - 수직 방향 

        - padding, margins, borders가 적용되지만 다른 요소를 밀어낼 수는 없다.

    - 수평 방향

        - padding, margins, borders가 적용되어 다른 요소를 밀어낼 수 있다. 

    - 이미지도 인라인 요소이지만, width나 height로 크기를 조정할 수 있다.

##### Block Content : 오른쪽의 남은 영역을 모두 차지한다. (상-하 방향)

    - h'' , p, div

    - 항상 새로운 행으로 나뉨

    - width와 height 속성을 사용하며 너비와 높이를 지정할 수 있다.

    - 기본적으로 width 속성을 지정하지 않으면 박스는 inline 방향으로 !!사용가능한 공간을 모두 차지한다!!.(너비를 사용가능한 공간의 100%로 채우는 것)


<br>

|/|block|inline|
|----|---|----|
|right|margin-right: auto;|text-align: left;|
|left|margin-left: auto;|text-align: right;|
|center|margin-right: auto;, margin-left: auto;|text-align: center;|


<br>

'inline-block' : inline과 block 요소 사이의 중간 지점을 제공하는 display 값

    - 사용 명령어 : "display: inline-block;"

    - block 요소의 특징을 가짐

        - width, height 속성 사용 가능

        - padding, margi, border로 인해 다른 요소가 밀려남

    - 요소가 줄바꿈 되는 것을 원하지 않으면서(Inline) 너비와 높이를 적용하고 싶은 경우(Block)에 사용

<br>

'None' : 요소를 화면에 표시하지 않고, 공간조차 부여되지 않는다.

    - 코드에는 포함이 되지만 화면에는 나타나지 않는다.

    - 추후 JS와 연계할 때 행동을 부여하는 역할로 사용된다.

- visibility: hidden; - 공간은 차지하지만 시각적으로 숨김 기능을 제공한다.

=> None과의 차이점 : 공간을 없애는지 안없애는지.

## CSS Layout Position

- 요소를 Normal Flow에서 제거하여 다른 위치로 배치하는 것

- 단순 선언만 하는 것이므로 이동을 하지 않는다면 현재 위치와 같다.

Position 이동 방향 : 동서남북상하

top, bottom, left, right를 사용하여 위치를 조절

<br>

- position 종류

    1. position: static; 
    
        - 가장 기본적인 position

        - 요소를 normal flow에 따라 배치

    2. position: relative;

        - 상대위치. 
        
        - 본인의 static(자기 자신) 위치를 기준으로 이동.
        
        - 본인의 과거 영역을 버리지 않음

    3. position: absolute;
    
        -절대위치. (집나간 친구)
        
        - static이 아닌 부모를 찾아간다.(가장 가까운 relative 부모 요소를 기준으로 이동) 
        
        - 자신이 차지하고 있던 영역을 버린다. 
        
        - 아래에 있던 영역들이 버린 영역을 채우게 된다.

        - 기준점을 명확하게 설정해 주어야 한다.

        - 레이아웃과 무관하지만 고정된 영역을 표현하는데 주로 사용된다.

    4. position: fixed;

        - 고정위치. (집나간 자식)

        - 화면에 보이는 브라우저에 고정되게 된다.(viewport를 기준으로 이동)

        - 문서에서 요소가 차지하는 공간이 없어진다.

        - ex) 네비게이션 / 웹툰 창 등

    5. position: sticky;

        - 임계점에 도달하기 전에는 일반적으로 동작(Normal flow를 따른다.)

        - 요소가 일반적인 문서 흐름에 따라 배치되다가 스크롤이 특정 임계점에 도달하면 그 위치에서 고정된다.
        
            - sticky -> fixed처럼 동작

        - 다른 sticky가 있을 경우 바통터치

    6. z-index

        - 정수 값을 사용해 z축 순서를 지정

        - 더 큰 값을 가진 요소가 작은 값의 요소를 덮음

<br>

## CSS Layout Flexbox

- 요소를 행과 열 형태로 배치하는 1차원 레이아웃 방식

--> **공간 배열 & 정렬**

- FlexBox는 부모(Flex Container)가 기준이 된다.(자식 : Flex Item)

<br>

#### 구성요소

- main axis : 주 축

    - flex item들이 배치되는 기본 축

    - main start에서 시작하여 main end 방향으로 배치

- cross axis : 교차 축

    - main axis에 수직인 축

    - cross start에서 시작하여 cross end 방향으로 배치

    - main 축만 기억하면 된다.

- Flex Container

    - **display : flex;** 혹은 display: inline-flex;가 설정된 부모 요소

    - 이 컨테이너의 1차 자식 요소들이 flex item이 된다

    - flexbox 속성 값들을 사용하여 자식 요소 flex item들을 배치

<br>

### layout 구성 요소

#### Container

- flex-direction: ''; - main 축을 행으로 할 것 인지, 열로 할 것인지 설정

    - row, column, row reverse, column reverse

- flex-wrap: ''; - 줄어들 때 감쌀 것인지, 감싸지 않을 것인지 판단

    - nowrap(기본값), wrap

- 주 축 정렬

    - justify-content: ''; - main 축 기준으로 정렬에 사용

        - flex-start(왼쪽 정렬), center(가운데 정렬), flex-end(오른쪽 정렬)

- 교차 축 정렬

    - align-content: ''; - 교차 축 정렬(flex-wrap을 해주어야 사용 가능)

        - flex-start, center, flex-end

    - align - items: ''; - 행 하나 정렬

        - flex-start, center, flex-end
<br>

#### item

- alig-self: ''; - 아이템 하나(or 자기 자신)에 대해서 정렬하고자 할 때

    - flex-start, center, flex-end
