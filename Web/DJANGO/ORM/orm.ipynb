## QuerySet API

: ORM에서 데이터를 검색, 필터링, 정렬 및 그룹화 하는데 사용하는 도구

-> API를 사용하여 SQL이 아닌 **python 코드**로 데이터를 처리

![Alt text](image-1.png)

- 구조 : `(Model Class).(Manager).(Queryset API(method))`

    ex) `Article.objects.all()`

    - method를 사용하려면 Manager가 필요하다.

#### Query

- 데이터베이스에 특정한 데이터를 보여 달라는 요청

- "쿼리문을 작성한다"

    -> 원하는 데이터를 얻기 위해 데이터베이스에 요청을 보낼 코드를 작성한다

- 파이썬으로 작성한 코드가 ORM에 의해 SQL로 변환되어 DB에 전달도며,  DB의 응답 데이터를 ORM이라는 `QuerySet`이라는 자료 형태로 변환하여 우리한테 전달.

##### QuerySet

- 데이터베이스에게서 전달받은 객체 목록(데이터 모음)

    - ***순회가 가능***한 데이터로써 1개 이상의 데이터를 불러와 사용할 수 있음

- Django ORM을 통해 만들어진 자료형

- 단, DB가 단일한 객체를 반환할 떄는 QuerySet이 아닌 모델(Class)의 인스턴스로 반환됨

### !!! Python의 모델 클래스와 인스턴스를 활용해 DB에 데이터를 저장, 조회, 수정, 삭제하는 것 !!!

## QuerySet API 실습

### Django shell

: Django 환경 안에서 실행되는 python shell(입력하는 QuerySet API 구문이 Django 프로젝트에 영향을 미침)

### Create

- 필요 외부 라이브러리 : ipython, django-extensions

    - django-extensions는 INSTALLED_APP에 추가해주어야 한다(써드파티 앱 확인해보기)

    - 장고 INSTALLED_APPS 권장 순서
        ```python
        INSTALLED_APPS = [
            # normal apps
            'articles',
            # third-party apps
            'django_extensions'
            # django apps
            'django.contrib.admin',
            'django.contrib.auth',
                    ...
        }
        ```

- `python manage.py shell_plus`

- 데이터 객체를 만드는(생성하는) 3가지 방법

    1.  ```
        article = Article() : Article(class)로부터 article(instance) 생성

        article.title = "titel이름" : 인스턴스 변수(title)에 값을 할당
        article.content = "content 내용" : 인스턴스 변수(content)에 값을 할당
        
        article.save() : 객체를 데이터베이스에 저장하는 메서드

        ```

        - articles은 배열의 형태를 띈다

            - `articles[0] : <Article: Article object (1)>`

        - 출력 방법

            - 클래스 이름.클래스 객체

            - article.pk : id 값 확인

    2. `article = Article(title='second', content='django!')`

        - 무조건 article.save()를 까먹으면 안된다.

    3. QuerySet API 중 create() 메서드 활용

        `Article.objects.create(title='third', content='django!')`
        
        - 위 2가지 방법과 달리 바로 저장 이후 바로 생성된 데이터가 반환된다. 

        - 단순히 변수일 뿐이므로 pk가 존재하지 않는다.

<br>

### Read

- all() : 전체 데이터 조회

    - `Article.objects.all()`

- get() : 단일 데이터 조회

    - 객체를 찾을 수 없으면 DoesNotExist 예외를 발생시키고, 둘 이상의 객체를 찾으면 MultipleObjectsReturned 예외를 발생시킴

    - `Article.objects.get(pk = '')`

    - get()은 pk에만 적합한 메서드이다. : pk는 데이터를 구분할 수 있는 유일한 값
            
- filter() : 특정 조건 데이터 조회

    ex) `Article.objects.filter(content='django!')` : content가 django!인 데이터들을 찾는다.

    - 조건을 만족하지 않아도 에러를 반환하지 않는다.(빈 결과 반환)

        -> 결과와 상관없이 무조건 QuerySet을 반환한다.

<br>

### Update

- 데이터 수정

    - 인스턴스 변수를 변경 후 save 메서드 호출

    ```
    article = Article.objects.get(pk=1) : 수정할 인스턴스 조회

    article.title = 'byebye' : 인스턴스 변수를 변경

    article.save() : DB에 저장
    ```

<br>

### Delete

- 데이터 삭제

    - 삭제하려는 데이터 조회 후 delete 메서드 호출

    - 잘 선택하고 잘 지우면 끝이다.

    `
    article = Article.objects.get(pk=1) : 삭제할 인스턴스 조회

    article.delete() " : delete 메서드 호출(삭제된 첫번째 객체가 반환)

    Article.objects.get(pk=1) : 삭제가 잘 되었는지 확인
    `

    - 이미 삭제한 DB 복구는 불가능하다.

[참고] 

- pk number는 재사용이 불가능하다.

- Field lookups

    - 특정 레코드에 대한 조건을 설정하는 방법

    - QuerySet 메서드 filter(), exclude() 및 get()에 대한 키워드 인자로 지정된다.

    - ex) `Article.objects.filter(content__contains='')`
