## Django and Framework

Django : 파이썬 기반의 대표적인 웹 프레임워크

### 클라이언트와 서버

```
웹의 동작 방식

클라이언트 --> 서버 : 요청

클라이언트 <-- 서버 : 응답
```

##### 클라이언트 : 서비스를 요청하는 주체( 웹 사용자의 인터넷이 연결된 장치, 웹 브라우저)

##### 서버 : 클라이언트의 요청에 응답하는 주체(웹 페이지, 앱을 저장하는 컴퓨터)

웹 페이지를 보개 되는 과정

1. 웹 브라우저에서 주소를 입력

2. 브라우저는 인터넷에 연결된 주소 컴퓨터(서버)에게 홈페이지 html 파일을 달라고 요청

3. 요청을 받은 주소 컴퓨터는 db에서 html 파일을 찾아 응답

4. 전달받은 html 파일을 웹 브라우저가 사람이 볼 수 있도록 해석해주면서 사용자는 주소의 메인 페이지를 보게 됨

**==> Django를 사용해서 서버를 구현할 것**

### 프로젝트 및 가상환경

가상환경 : Python 애플리케이션과 그에 따른 패키지들을 *격리*하여 관리할 수 있는 독립적인 실행 환경

##### `python -m venv venv(마지막 venv는 가상환경 이름)` : 가상환경 생성

! 하나의 터미널에서는 하나의 가상환경만 사용 가능

! 가상환경 2개를 독립적으로 사용하고 싶다면 새로운 터미널을 통해 오픈해야 한다.

가상환경 활성화 : `source venv/Scripts/activate(window 기준)`

가상환경 비활성화 : `deactivate`

<br>

- 가상환경의 패키지 목록이 필요한 경우 text file로 만들어 공유할 수 있다.

    - 명령어 : `pip freeze > requirements.txt`

    - 패키지 목록 저장 후 새로운 패키지를 설치할 경우 자동으로 업데이트 되지 않으므로, 새로 `pip freeze > requirements.txt` 를 해주어야 한다.

- 저장한 패키지 text file을 사용하여 새로운 환경에 설치할 수 있다.

    - 명령어 : `pip install -r 텍스트파일.txt`

[참고]

- 가상환경 폴더는 원격 저장소(Github 등)에 올리지 않는다.

    - gitingore 에 업로드 할 것

    - 가상환경 이름, 패키지파일 텍스트 이름 은 개발자들의 공용이므로 이름 그대로 사용하는 것을 권장함

#### Django 프로젝트 생성 전 루틴

1. 가상환경 생성

2. 가상환경 활성화

3. Django 설치

4. 의존성 파일 생성(패키지 설치시마다 진행) : 파이썬 패키지 txt 파일

5. .gitignore 파일 생성(첫 add 전)

6. git 저장소 생성

7. Django 프로젝트 생성

<br>

- Django 프로젝트 생성 : `django-admin startproject '프로젝트이름' .`

- Django 서버 실행 : `python manage.py runserver`

## Django project

- 애플리케이션의 집합(DB설정, url 연결, 전체 앱 설정 등을 처리)

    - Django application : 독립적으로 작동하는 기능 단위 모듈

        - 각자 특정한 기능을 담당하며 다른 앱들과 함께 하나의 프로젝트를 구성

- 앱 생성 : `python manage.py startapp articles` (앱의 이름은 복수형으로 지정하는 것을 권장)

- 앱 등록 : 생성한 앱은 projects -> settings.py의 INSTALLED_APPS 배열 위쪽부터 작성할 것

## Django 디자인 패턴

: 소프트웨어 설계에서 발생하는 문제를 해결하기 위한 일반적인 해결책(공통적인 문제를 해결하는데 쓰이는 형식화된 관행)

<br>

### MVC 디자인 패턴(Model, View, Controller)

: 애플리케이션을 구조화하는 대표적인 패턴(데이터, 사용자, 인터페이스, 비즈니스 로직을 분리)

    - 서로 영향을 받지 않고, 독립적이고, 쉽게 유지.보수할 수 있는 애플리케이션 구조

    - Django에서는 `MTV 디자인 패턴(Model, Template, View)` 이라고 부른다.

#### 프로젝트 구조

- `settings.py` : 프로젝트의 모든 설정을 관리

- `urls.py` : URL과 이에 해당돠는 적절한 views를 연결
