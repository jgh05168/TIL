## Static Files

: 정적 파일, 서버 측에서 변경되지 않고 고정적으로 제공되는 파일(이미지, JS, CSS파일 등)

웹 서버의 기본 동작 : 특정 위치(url)에 있는 자원을 요청받아서 응답을 처리하고 제공하는 것

--> **정적 파일을 제공하기 위한 경로(url)**가 있어야 한다.

![Alt text](image.png)

[Static files 기본 경로] : app폴더/static/

- static file 은 static tag를 사용해서 불러온다.

- {% load static %}을 정의해 주어야만 static tag를 사용할 수 있다.

### STATIC_URL

: 기본 경로 및 추가 경로에 위치한 정적 파일을 참조하기 위한 URL

- 실제 파일이나 디렉토리가 아니며, URL로만 존재.

```
# settings.py

STATIC_URL = 'static/
```

- URL + *STATIC_URL* + **정적파일 경로**

    ex) http://127.0.0.1:8000/*static*/**articles/sample-1.png**

### Static files 추가 경로

: STATICFILES_DIRS에 문자열 값으로 추가 경로 설정

    - STATICFILES_DIRS는 리스트로 저장한다.

## Media Files

: 사용자가 웹에서 업로드하는 정적 파일(user-uploaded)

### ImageField()

: 이미지 업로드에 사용하는 모델 필드

- 이미지 객체가 직접 저장되는 것이 아닌 '이미지 파일의 경로'가 문자열로 DB에 저장

[사전 준비]

1. settings.py에 MEDIA_ROOT, MEDIA_URL 설정

2. 작성한 MEDIA_ROOT와 MEDIA_URL에 대한 url 지정


#### MEDIA_ROOT

: 미디어 파일들이 위치하는 디렉토리의 절대 경로(리스트가 아니다)

```
# settings.py

MEDIA_ROOT = BASE_DIR / 'media'
```

#### MEDIA_URL

: MEDIA_ROOT에서 제공되는 미디어 파일에 대한 주소를 생성(STATIC_URL과 동일한 역할)

```
# settings.py

MEDIA_URL = 'media/'
```
