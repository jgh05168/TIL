## Django Fixtures

: Django가 데이터베이스로 가져오는 방법을 알고 있는 데이터 모음

-> 데이터베이스 구조에 맞추어 작성 되어있음.

- fixtures를 사용해 앱에 초기 데이터(initial data)를 제공

### `dumpdata`

: 데이터베이스의 모든 데이터를 추출 -> 추출한 데이터는 json 형식으로 저장

ex) python manage.py dumpdata --indent 4 articles.article > articls.json

### `loaddata`

: Fixtures 데이터를 데이터베이스로 불러오기

[Fixtures 파일 기본 경로] : app_name/fixtures

-> Django는 설치된 모든 app의 디렉토리에서 fixtuers 폴더 이후의 경로로 fixtures 파일을 찾아 load

ex) python manage.py loaddata articles.json

[loaddata 순서 주의사항]

    - 만약 loadddata를 한번에 실행하지 않고 하나씩 실행한다면 모델 관계에 따라 load하는 순서가 중요할 수 있음

        - coment는 article에 대한 key 및 user에 대한 key 가 필요

        - article은 user에 대한 key가 필요

!!!! Fixtures 파일을 직접 만들지 말 것 !!!!

- 반드시 dumpdata 명령어를 사용하여 생성

## Improve Query

: 같은 결과를 얻기 위해 DB 측에 보내는 쿼리 개수를 점차 줄여 조회하기

### annotate

: SQL의 GROUP BY 쿼리를 사용

문저 : 각 게시글마다 댓글 개수를 반복 평가

해결 : 게시글을 조회하면서 댓글 개수까지 한번에 조회해서 가져오기

ex) `articles = Article.objects.annotate(Count('comment')).order_by('-pk')`

### select_related

: SQL의 INNER JOIN 쿼리를 활용 -> 1:1 또는 N:1 참조 관계에서 사용

문제 : 각 게시글마다 작성한 유저명까지 반복 평가

해결 : 게시글을 조회하면서 유저 정보까지 한번에 조회해서 가져오기

ex) `articles = Article.objects.select_related('user').order_by('pk')`
