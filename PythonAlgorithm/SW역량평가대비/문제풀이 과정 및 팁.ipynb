### 문제풀이 과정

1. 문제 설계

    1) 문제 해석

        - 문제에서 원하는 목표 -> 명확하게 파악해야 한다.

        - 만약 문제가 이해가 가지 않는다면 문제 내 조건들을 하나씩 적어보기

    2) 알고리즘 선택

        - 문제 해석과는 별개로 생각한다.

        - 내 해석과 문제에서 원하는 출력이 같은지 확인

        - 머리 속으로 시뮬레이션을 해본다.

            - 조건들을 모두 생각해보며 가능할까 하는 경우도 모두 생각해 보아야 한다.
        
            ex) 모든 경우를 탐색해 봐야 한다면, 완전 탐색

        - 이후, 떠오르는 알고리즘이 존재한다면 **시간복잡도를 계산**해본다.

            - 그리디(규칙 찾기), dp(두 번 계산하지 않기), union-find, 탐색알고리즘 등등

            !! *코딩테스트에서 파이썬은 초당 2000만번의 연산을 수행한다.* !!

            ex) 완전 탐색으로 시간복잡도를 계산 했을 때 시간초과가 날 것 같다 ?

                -> 백트래킹 + 가지치기 사용

        - Test Case를 사용하여 선택한 알고리즘으로 문제가 풀리는지 시뮬레이션
