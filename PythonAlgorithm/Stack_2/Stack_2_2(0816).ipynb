## 부분 집합의 합



```python
# 재귀함수 연습 -- A 배열의 내용을 B에 복사하기
def f(i, N):
    if i == N:
        return
    else:
        B[i] = A[i]
        f(i + 1, N)
        return

N = 3
A = [1, 2, 3]
B = [0] * N
print(A, B)
f(0, N)
print(A, B)
```

```python
# N에 대한 부분집합 구하기

def f(i, N):
    if i == N:
        print(bit, end=' ')
        for j in range(N):
            if bit[j]:
                print(A[j], end=' ')
        print()
        return
    else:
        bit[i] = 1
        f(i + 1, N)
        bit[i] = 0
        f(i + 1, N)

A = [1, 2, 3]
bit = [0] * 3
f(0, 3)
```

```python
# 부분집합의 합 구하기
def f(i, N):
    if i == N:
        print(bit, end=' ')
        s = 0               # 부분집합의 합
        for j in range(N):
            if bit[j]:
                s += A[j]
                print(A[j], end=' ')
        print(f' : {s}')
        return
    else:
        bit[i] = 1
        f(i + 1, N)
        bit[i] = 0
        f(i + 1, N)

A = [1, 2, 3]
bit = [0] * 3
f(0, 3)
```

```python
# s : i - 1 원소까지 부분집합의 합(포함된 원소의 합)
def f(i, N, s):
    if i == N:
        print(bit, end=' ')
        print(f' : {s}')
        return
    else:
        bit[i] = 1      # 부분집합에 A[i] 포함
        f(i + 1, N, s + A[i])
        bit[i] = 0      # 부분집합에 A[i] 빠짐
        f(i + 1, N, s)  # 이전까지 한 과정과 같음

A = [1, 2, 3]
bit = [0] * 3
f(0, 3, 0)
```

```python
def Power_2(Base, Exp):
    global cnt2
    if Exp == 0 or Base == 0:
        return 1
    
    # 홀수인 경우
    if Exp % 2:
        result = Power_2(Base, (Exp - 1) // 2)
        cnt2 += 1
        return result * result * Base
    # 짝수인 경우
    else:
        result = Power_2(Base, Exp // 2)
        cnt2 += 1
        return result * result

cnt2 = 0
print(Power_2(2, 10), cnt2)
```
