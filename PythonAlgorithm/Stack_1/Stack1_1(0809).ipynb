## 스택

- 자료를 쌓아 올린 형태의 자료구조

- **선형 구조를 갖는다**

    선형 구조 : 자료 간의 관계가 1 대 1

    비선형 구조 : 자료 간의 관계가 1 대 N

- **LIFO(Last In First Out)** : 마지막에 삽입한 자료를 가장 먼저 꺼낸다 - 후입선출

    ex) 
    
    input : 1 2 3

    output : 3 2 1

- 스택의 top에 있는 item을 반환하는 연산(peak)

#### push

```python
def push(s, item):

```

````python
# append를 쓰지 않는 stack
def push(item, size):
    global top
    top += 1
    if top == size:
        exit()
    else:
        stack[top] = item

size= 10
stack = [0] * size
top = -1

push(10, size)

# push(20) - 간단하게 작성
top += 1
stack[top] = 20
````

## 재귀 호출

- 자기 자신을 호출하여 순환 수행되는 것

- 함수에서 실행해야 하는 작업의 특성에 따라 일반적인 호출 방식보다 재귀호출방식을 사용하여 프로그램의 크기를 줄인다

ex) 팩토리얼을 재귀함수로



```python
# ex) fibonacci

def fibo(n):
    if n < 2:
        return n
    else:
        return fibo(n - 1) + fibo(n - 2)
```

## 메모아이제이션(Memoization)

- 재귀 함수의 중복 호출을 방지하기 위한 방법

- 이전에 계산한 값을 메모리에 저장해서 다시 계산하지 않도록 한다

- 실행속도를 빠르게 함


```python
# fibo를 memoization으로
def fibo1(n):
    global memo
    if n >= 2 ad memo == 0:
        memo[n] = (fibo1(n - 1) + fibo1(n - 2))
    return memo[n]

memo = [0] * (n + 1)
memo[1] = 1 
```
