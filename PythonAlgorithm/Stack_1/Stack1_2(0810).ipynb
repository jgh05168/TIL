## DP

- 동적 계획법 : 최적화 문제를 해결하는 알고리즘

- 크기가 작은 부분 문제들을 모두 해결한 후, 그 해들을 이용하여 큰 크기의 부분 문제들을 해결하여 최종 문제를 해결하는 방식

```python
# 피보나치 예제

# 1. 재귀함수
def fibo2(n):
    f = [0] * (n + 1)
    f[0] = 0
    f[1] = 1
    for i in range(2, n + 1):
        f[i] = f[i - 1] + f[i - 2]

    return f[n]
```

## DFS : 깊이 우선 탐색

- 시작 정점의 한 방향으로 갈 수 있는 경로까지 간다.

- 더이상 갈 수 없으면 정점을 되돌아가 다른 방향의 정점을 탐색

- 모든 정점을 반복하면 종료

### 순서

1. 시작 정점 v를 결정하여 방문한다.

2. 정점 v에 인접한 정점 중에서

    2-1. 방문하지 않은 정점 w가 있으면 v를 스택에 push한 뒤 정점 w를 방문

    2-2. 방문하지 않은 정점이 없으면, 탐색의 방향을 바꾸기 위해 스택을 pop한 뒤 가장 마지막 방문 정점을 v로 하여 다시 2. 를 반복

3. 끝날때까지 반복한다.


```python
visited = [False] * n
stack = []

def DFS(v):
    visited[v] = True
    while True:
        if v의 인접 정점 중 방문 안 한 정점 w가 있으면:
            push(v)
            v <- w
            visited[w] = True
        else:
            if stack이 비어있지 않다면:
                v = stack.pop
            else:
                break

```

! 조건을 명확하게 주지 않는다면 DFS 의 탐색 순서는 여러가지가 될 수 있다.




DFS 구현



````python
# ex) DFS

def DFS(n, V, adj_matrix):
    stack = []
    visited = [False] * (V + 1)
    visited[n] = True
    print(n)
    while 1: 
        for w in range(1, V + 1):           
            if adj_matrix[n][w] == 1 and visited[w] == False:   # 현재 정점 n에 인접하고, 미방문
                stack.append(n)
                n = w
                print(n)
                visited[n] = True
                break               # for w, n에 인접인 w를 찾은 경우
        else:
            if stack:               # 스택에 값이 존재한다면, 단순 값 pop
                print(1)
                n = stack.pop()
            else:                   # 스택이 비었다면
                break               # dfs 종료


V, E = map(int, input().split())    # 1번부터 V번 정점, E개의 간선의 개수
arr = list(map(int, input().split()))
adj_mtrix = [[0] * (V + 1) for _ in range(V + 1)]   # 빈 인접행렬 생성

for i in range(E):
    v1, v2 = arr[i * 2], arr[i * 2 + 1]
    # 대칭되는 행렬을 생성한다
    adj_mtrix[v1][v2] = 1
    adj_mtrix[v2][v1] = 1


DFS(1, V, adj_mtrix)
````
