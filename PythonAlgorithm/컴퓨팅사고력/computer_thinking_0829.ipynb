## 프로그래밍과 논리/수학

#### [참고]

- 명제(p, q, r, ...) : 참이나 거짓을 알 수 있는 식이나 표현

- 진리값(T / F 또는 0 / 1)

- NOT(! 또는 ~) : 진리값이 반대

- AND(p^q)

- OR(pVq)

- XOR(⊕) : p, q가 명제일 때, p, q 중 하나만 참일 때 참이 되는 명제

[합성]

- 항진명제 : 진릿값이 항상 참

- 모순명제 : 진리값이 항상 거짓

- 사건명제 : 항진명제도 모순명제도 아닌 명제

- 조건명제

    - p, q가 명제일 때, 명제 p가 조건, q가 결론으로 제시되는 명제

    - p -> q : p이면 q이다

    

- 쌍방조건명제

    - p, q가 명제일 때, 명제 p와 q가 모두 조건이면서 결론인 명제

    - p면 q고, q면 p이다.



<br>

증명은 정확한 명제식으로 표현할 수 있는 것이여야 한다.

- Trivial Proof: 모든 x에 대해 P(x) -> Q(x)를 증명하려는데, Q(x)가 항상 참인 경우

    -> Q(x)가 참(T)인 것만을 확인한다면 조건명제는 참이다! 를 증명할 수 있다.

- Vacuous Proof: 모든 x에 대해 P(x) -> Q(x)를 증명하려는데, P(x)가 항상 거짓인 경우
 
    -> P(x)가 가짓(F)인 것만을 확인한다면 조건명제는 참이다! 를 증명할 수 있다.

<br>

## 논리와 증명

- 모든 x 문제에 대해서 증명할 경우에는 반례를 잘 이용하여 참, 거짓을 판단하면 된다.

- 어떤 x 문제에 대해서는 해당 범위 내에 x가 하나라도 존재하면 된다.

<br>

## 수와 표현

어떤 값 n을 표현하기 위해서는 ?!

    - 2^k - 1 >= n 이 성립해야 한다.

    -> k >= log(n + 1) : 약 logn 비트가 필요함

! 컴퓨터 분야에서 log의 밑은 항상 2

log를 사용하는 이유 : 매우 큰 값을 작게 표현할 수 있다.

<br>

## 집합과 조합론

- 조합론은 보통 경우의 수를 따지는 문제들을 말한다.

    - 보통 C를 이용하여 표현하지만 괄호 표현을 더 자주 쓴다.

## 기초 수식

- 알고리즘의 시간복잡도를 계산할 수 있고, 예측이 가능하다.

ex)

1. 

T(n) = T(n - 1) + 1, T(0) = 1 을 O()로 풀어보시오.

     = T(n - 2) + 1 + 1

     = T(n - 3) + 1 + 1 + 1

              ....

     = T(n - k) + k , if n == k

     = 1 + n = O(n)

2. 

T(n) = T(n - 1) + n, T(0) = 1 을 O()로 풀어보시오.

     = T(n - 2) + (n - 1) + n

     = T(n - 3) + (n - 2) + (n - 1) + n

              ....

     = T(n - k) + (n - k + 1) + ... + n , if n == k

     = 1 + 2 + ... + (n - 1) + n = n(n + 1) / 2

     = O(n^2)
