## 2차원 배열

- 1차원 List를 묶어놓은 list

- 2차원 이상의 다차원 list는 차원에 따라 index 선언

- 행(row) x 열(col)

```python
arr = [[1, 2, 3], [4, 5, 6]]        # row = 0, 1  /  col = 0, 1, 2

```

또는

```python
N = int(input())
arr = [list(map(int, input().split())) for _ in range(N)]
# 보통 2차원 배열을 입력받는 방법

```
<br>

### 배열 순회

- n x m 배열의 n * m 개의 모든 원소를 빠짐없이 조사하는  방법

<br>

#### 행 우선 순회

```python
# i 행의 좌표
# j 열의 좌표

for i in range(n):
    for j in range(m):
        print(arr[i][j])
```

#### 열 우선 순회

```python
# i 행의 좌표
# j 열의 좌표

for j in range(m):
    for i in range(n):
        print(arr[i][j])
```

#### 지그재그 순회

```python
# i 행의 좌표
# j 열의 좌표
# 짝수 행은 오른쪽으로, 홀수 행은 왼쪽으로

# 1. 
for i in range(n):
    for j in range(m):
        # 짝수 행일 때 : arr[i][j] | 홀수 행일 때 : arr[i][m - 1 - j]
        print(arr[i][j + (m - 1 - 2 * j) * (i % 2)])
        # i % 2로 짝수 / 홀수 행 구분 => 짝수일 경우 통쨰로 날려주고 홀수일 경우 m-1-j 꼴로 만들어줌
        # j + (m-1-2*j) 부분은 m-1-j와 동일하다 

# 2. 
for i in range(n):
    if i % 2 == 0:
        for j in range(m):
            print(arr[i][j])
    else:
        for j in range(m - 1, -1, -1):
            print(arr[i][j])
```

#### 델타를 이용한 2차 배열 탐색 

#### (!!!!!!!!!!!! 가장 중요 !!!!!!!!!!!!!)

- 2차원 배열의 한 좌표에서 4방향(상, 하, 좌, 우)의 인접 배열 요소를 탐색하는 방법

- (0, 1) : right | (1, 0) : down | (0, -1) : left | (-1, 0) : up

```python
drow = [0, 1, 0, -1]
dcol = [1, 0, -1, 0]

for i in range(N):
    for j in range(N):
        for k in range(4):
            ni = i + drow[k]
            nj = j + dcol[k]
            if 0 <= ni and N > ni and 0 <= nj and N > nj:       # 유효한 인덱스인지 확인하는 과정
            print(arr[ni][nj])

```

#### 전치 행렬(Transpose)

```python
# i : 행의 좌표, len(arr)
# j : 열의 좌표, len(arr[0])

for i in range(N):
    for j in range(N):
        if i < j:           # 자리를 두 번 바꾸는 행위를 막는 조건
            ar[i][j], arr[j][i] = arr[j][i], arr[i][j]
```
