## 조합

- 서로 다른 n개의 원소 중 r개를 순서없이 골라낸 것

- nCr

```python
an[] : n 개의 원소를 가지고 있는 배열
tr[] : r 개의 크기의 배열, 조합이 임시 저장된 배열

comb(n, r)
    if (r == 0): print_arr()
    elif (n < r): return
    else:
        tr[r - 1] = an[n - 1]
        comb(n - 1, r - 1)
        comb(n - 1, r)
```

````python
# ex1) 조합(5C3) - 뒤에서부터 만들어져 출력된다.
def ncr(n, r):
    if r == 0:
        print(tr)
    elif n < r:         # 남은 원소보다 많은 원소를 선택해야하는 경우
        return          # 불가능
    else:
        tr[r - 1] = a[n - 1]        # a[n - 1] 조합에 포함시키는 경우
        ncr(n - 1, r - 1)           
        ncr(n - 1, r)               # a[n - 1]을 포함시키지 않는 경우

a = [1, 2, 3, 4, 5]
n = 5
r = 3
tr = [0] * r
ncr(n, r)
````

## Greedy Algorithm

- 최적해를 구하는 데 사용되는 근시안적 방법

- 여러 경우 하나를 선택 할 때마다 그 순간에 최적이라고 생각되는 것을 선택해 나가는 방식

- **각 선택에 대해 최적의 선택을 하지만, 최종 결과가 최적의 값이 될 것이라는 보장이 없다.**

<br>

#### 탐욕 알고리즘의 동작 과정

1) 해 선택 : 현재 상태에서 부분 문제의 최적 해를 구한 뒤, 이를 부분해집합에 추가한다.

2) 실행 가능성 검사 : 새로운 부분 해 집합이 실행가능한지, 문제 제약조건을 위반하지 않는지를 확인한다.

3) 해 검사 : 새로운 부분 해 집합이 문제의 해가 되는지 확인한다. 아직 전체 문제의 해가 완성되지 않았다면, 1의 해부터 다시 시작.

## 활동 선택 문제

- 시작 시간과 종료 시간이 있는 n 개의 활동들의 집합에서 서로 겹치지 않는 최대 갯수의 활동들의 집합 s를 구하는 문제

- 양립 가능한 활동들의 크기가 최대가 되는 s(0, n + 1)의 부분집합을 선택하는 문제

    1. 종료 시간 순으로 활동들을 정렬

#### --> 탐욕 기법의 적용

1) 하위 문제 Sij에서 종료시간이 가장 빠른 활동 am을 선택

2) Sim을 선택하게 되면 공집합이 아닌 하위문제 Smj가 남게된다.

3) 1, 2 반복

```
A : 활동들의 집합, S : 선택된 활동들 집합
si : 시작시간, fi : 종료시간, 1 <= i <= n

A.sort()
S <- [A, ]                  # 첫번째 활동 선택
j <- 1
for i in (1, n)
    if si >= fi
        s <- S.union(Ai)    # 선택한 활동의 종료시간보다 빠른 시작시간을 가지는 활동을 모두 제거
        j <- i
```

````python
N = 10
a = [1, 4, 1, 6, 6, 10, 5, 7, 3, 8, 5, 9, 3, 5, 8, 11, 2, 13, 12, 14]

meet = []
for i in range(N):
    meet.append([a[i * 2], a[i * 2 + 1]])
meet.sort(key=lambda x:x[1])
meet = [[0, 0]] + meet
S = [1]
j = 0

for i in range(1, N + 1):
    if meet[i][0] >= meet[j][1]:
        S.append(i)
        j = i
print(S)
````

#### 탐욕 알고리즘의 필수 요소

- 탐욕적 선택 속성(greedy choice property)

    - 탐욕적 선택은 최적해로 갈 수 있음을 보여라

    -> 즉, 탐욕적 선택은 항상 안전하다.

- 최적 부분 구조

    - 최적화 문제를 정형화

    -> 하나의 선택을 하면 풀어야 할 하나의 하위 문제가 남는다

**[원문제의 최적해 = 탐욕적 선택 + 하위 문제의 최적해]**

|탐욕 기법|동적 계획법|
|:------|:---|
|매 단계에서, 가장 좋게 보이는 것을 빠르게 선택한다. -> 지역 최적 선택|매 단계의 선택은 해결한 하위 문제의 해를 기반으로 한다.|
|하위 문제를 풀기 전에 (탐욕적) 선택이 먼저 이루어진다.|하위 문제가 우선 해결된다.|
|Top-down|Bottom-up|
|빠르고, 간결하다|느리고, 복잡하다|
