## 클래스

클래스란? 파이썬에서 타입을 표현하는 방법

    -> 객체를 생성하기 위한 설계도

    -> 데이터와 기능을 함께 묶는 방법을 제공


### 생성자 함수(def _ _ init _ _(self))

    - 객체를 생성할 때 자동으로 호출되는 특별한 메서드

    - __init__으로 정의되며, 객체의 초기화를 담당

    - 생성자 함수를 통해 인스턴스를 생성하고 필요한 초기값을 설정

### 인스턴스 변수(self."")

    - 인스턴스(객체)마다 별도로 유지되는 변수

    - 인스턴스마다 독립적인 값을 가지며, 인스턴스가 생성될 떄마다 초기화

### 클래스 변수

    - 클래스 내부에 선언된 변수

    - 클래스로 생성된 모든 인스턴스들이 공유하는 변수

### 인스턴스 메서드

    - 각각의 인스턴스에서 호출할 수 있는 메서드

    - 인스턴스 변수에 접근하고 수정하는 등의 작업 수행
    

````python
class Person:
    # 속성(변수)
    blood_color = 'red'

    # 메서드

    # 매직메서드(개발자가 직접 호출하지 않음)
    def __init__(self, name):           # __init__ : 생성자 메서드
        self.name = name                # name : 위치 인자

    def singing(self):
        return f'{self.name}가 노래합니다.'
    
# 인스턴스 생성
singer1 = Person('iu')
singer2 = Person('BTS')

# 메서드 호출
print(singer1.singing())            # singing 뒤에 괄호 주의!!
print(singer2.singing())

### singer1과 singer2 는 독립적이다.

# 속성(변수) 사용       -->     같은 클래스를 참조했기 떄문에 결과가 같다.
print(singer1.blood_color)
print(singer2.blood_color)
````

인스턴스와 클래스 간의 이름 공간(namespace)



````python
# ex) (좋은 코드는 아님)

class Person:
    # 속성(변수)
    name = 'unknown'

    def talk(self):
        print(self.name)

p1 = Person()
p1.talk()       # unknown
# 인스턴스는 본인의 인스턴스 변수가 없으면 클래스 변수를 찾아 사용한다.

p2 = Person()
p2.talk()       # unknown

p2.name = 'Kim' # p2의 인스턴스 변수 정의
p2.talk()       # Kim

#!!! 인스턴스는 서로 독립적임 !!!#
print(Person.name)
print(p1.name)
print(p2.name)

p2.address = 'Korea'        # 새로운 인스턴스 변수를 생성해도 실행이 된다.
print(p2.address)       
````

! 독립적인 이름공간을 갖는 이점

    1. 각 인스턴스는 독립적인 '메모리 공간'을 가지며, 클래스와 다른 인스턴스 간에는 서로의 데이터나 상태에 직접적인 접근이 불가능함

    2. 객체 지향 프로그래밍의 중요한 특성 중 하나로, 클래스와 인스턴스를 모듈화하고 각각의 객체가 독립적으로 동작하도록 보장

    3. 클래스와 인스턴스는 다른 객체들과의 상호작용에서 서로 충돌에 영향을 주지 않으며 독립적으로 동작이 가능

    4. 코드의 가독성, 유지보수성, 재사용성을 높이는데 도움을 줌



````python
# ex) 클래스 변수 활용

class Person:
    count = 0

    def __init__(self, name):
        self.name = name
        Person.count += 1

p1 = Person('iu')
p2 = Person('BTS')

print(Person.count)
print(p1.count)
print(p2.count)

#### 클래스 변수를 변경할 때는 항상 "클래스.클래스변수" 형식으로 변경
#### "인스턴스.클래스변수" 로 변경할 경우 변경하는 인스턴스에서만 사용 가능
````

## 메서드

### 인스턴스 메서드

-> 클래스로부터 생성된 각 인스턴스에서 호출할 수 있는 메서드

-> 인스턴스의 상태를 조작하거나 동작을 수행

<br>

- 구조 

    ```python
    class MyClass:
        
        def instance_method(self, arg1, ...):
            pass
    ```

    - 클래스 내부에 정의되는 메서드의 기본

    - 반드시 첫 번째 매개변수로 인스턴스 자신(self)을 전달받음


````python
# ex)

# 인스턴스.메서드()     : 클래스.메서드()의 축약형 
'abc'.upper()

# 클래스.메서드(인스턴스 자기 자신)
str.upper('abc')
````

#### self 동작 원리

--> str 클래스가 upper 메서드를 호출했고, 그 첫번째 인자로 문자열 인스턴스가 들어간 것 

!!!!! 인스턴스 메서드의 첫번째 매개변수가 반드시 인스턴스 자기 자신인 이유

#### 생성자 메서드 (  _ _ init _ _())

-> 인스턴스 객체가 생성될 때 자동으로 호출되는 메서드

-> 인스턴스 변수들의 초기값을 설정

<br>

### 클래스 메서드

-> 클래스가 호출하는 메서드

-> 클래스 변수를 조작하거나 클래스 레벨의 동작을 수행

<br>

- 구조

    ```python
    class Myclass:

        @classmethod
        def class_method(cls, arg1, ...):
            pass        # cls.변수 = ""
    ```

    - @classmethod(**데코레이터**)를 사용하여 정의

    - 호출 시, 첫번째 인자로 호출하는 클래스(cls)가 전달됨

````python
# ex)

class Person:
    count = 0

    def __init__(self, name):
        self.name = name
        Person.count += 1

    @classmethod
    def number_of_population(cls):
        print(f"인구수 : {cls.count}")

p1 = Person('iu')
p2 = Person('BTS')

Person.number_of_population()
````

### 스태틱(정적) 메서드

-> 클래스와 인스턴스와 상관없이 독립적으로 동작하는 메서드

-> 주로 클래스와 관련이 있지만 인스턴스와 상호작용이 필요하지 않은 경우에 사용

<br>

- 구조

    ```python
    class MyClass:
        
        @staticmethod
        def static_method(arg1, ...):
            pass
    ```

    - @staticmethod 데코레이터를 사용하여 정의

    - 호출 시 필수적으로 작성해야 할 매개변수가 없음

    - 객체 상태나 클래스 상태를 수정할 수 없으며 단지 기능(행동)만을 위한 메서드로 사용

````python
class StringUtils:
    @staticmethod
    def reverse_string(string):
        return string[::-1]

    @staticmethod
    def capitalize_string(string):
        return string.capitalize()

text = 'hello, world'
reversed_txt = StringUtils.reverse_string(text)
print(reversed_txt)

capitalize_txt = StringUtils.capitalize_string(text)
print(capitalize_txt)
````

#### 메서드 정리

1. 인스턴스 메서드(self)

    - 인스턴스의 상태를 변경하거니, 해당 인스턴스의 특정 동작을 수행

<br>

2. 클래스 메서드(cls)

    - @classmethod 데코레이터

    - 인스턴스의 상태에 의존하지 않는 기능을 정의

    - 클래스 변수를 조작하거나 클래스 레벨의 동작을 수행

<br>

3. 스태틱 메서드

    - @staticmethod 데코레이터

    - 클래스 및 인스턴스와 관련이 없는 일반적인 기능을 수행

<br>

#### 각자의 역할

- 클래스가 사용해야 할 것 

    - 클래스 메서드

    - 스태틱 메서드

- 인스턴스가 사용해야 할 것

    - 인스턴스 메서드

<br>

- 클래스가 할 수 있는 것

    ```python
    instance = MyClass:

    print(MyClass.instance_metehod(instance))       # 이렇게 쓰지 말자
    print(MyClass.class_metehod())
    print(MyClass.static_metehod())
    ```

    ! 클래스는 모든 메서드를 호출할 수 있음

    !!!! 하지만 클래스는 클래스 메서드와 스태틱 메서드만 사용하도록 한다 !!!!

<br>

- 인스턴스가 할 수 있는 것

    ```python
    instance = MyClass:

    print(instance.instance_method())       
    print(instance.class_method())      # 이렇게 쓰지 말자
    print(instance.static_method())     # 이렇게 쓰지 말자
    ```

    ! 인스턴스는 모든 메서드를 호출 할 수 있음

    !!!! 하지만 인스턴스는 인스턴스 메서드만 사용하도록 한다 !!!!
